# Quality

1. Convert the "Genres" column from a string of lists to an actual list.

2. Transform the "Keywords" column, currently a string of lists, into a list of individual words.

3. Extract only the first three names from the "CAST" column.

4. Retain only the director's name from the "CREW" column.

5. Convert the "Overview" column, currently a string, into a list.

6. Remove spaces between names in all columns.

```python
def convert(obj):
    
# Initialize an empty list to store names
    L = []
    
# Safely convert the input string into a Python expression using ast(Abstract Syntax Tree).literal_eval
# The input string is expected to represent a list of dictionaries
    for i in ast.literal_eval(obj):
        
        # Append the 'name' key from each dictionary to the list
        L.append(i['name'])
        
    # Return the final list of names
    return L
```

```python
# The 'convert' function is designed to extract names from a string representation of a list of dictionaries
og_df['genres'] = og_df['genres'].apply(convert)
```

```python
og_df['keywords'].value_counts()
```

```python
# The 'convert' function is designed to extract names from a string representation of a list of dictionaries
og_df['keywords'] = og_df['keywords'].apply(convert)
```

```python
def convert3(obj):
    
    # Initialize an empty list to store names
    L = []
    
    # Initialize a counter to keep track of the number of names extracted
    counter = 0
    
    # In this case, the string represents a list of dictionaries
    for i in ast.literal_eval(obj):
        
        # Check if the counter is less than 3 to limit extraction to the first three names
        if counter != 3:
            
             # Append the 'name' key from each dictionary to the list
            L.append(i['name'])
            
             # Increment the counter
            counter+=1
        else:
            
            # Break the loop if three names have already been extracted
            break
            
    # Return the list of up to three names        
    return L
```

```python
# Apply the 'convert3' function to the 'cast' column of the 'og_df' DataFrame
og_df['cast'] = og_df['cast'].apply(convert3)
```

```python
def fetch_director(obj):
    
    # Initialize an empty list to store director names
    L = []
    
    # Use ast.literal_eval to safely evaluate the string as a literal Python expression
    for i in ast.literal_eval(obj):
        
        # Check if the 'job' key in the dictionary is 'Director'
        if i['job'] == 'Director':
            
             # Append the 'name' key from the dictionary to the list
            L.append(i['name'])
            
    # Return the list of director names
    return L
```

```python
# The 'fetch_director' function extracts director names from a string representation of a list of dictionaries in the 'crew' column
og_df['crew'] = og_df['crew'].apply(fetch_director)
```

```python
import nltk
```

```python
# Import the PorterStemmer class from the NLTK library
from nltk.stem.porter import PorterStemmer

# Create an instance of the PorterStemmer
ps = PorterStemmer()
```

```python
# Define a stemming function using the PorterStemmer from NLTK
def stem(text):
    # Initialize an empty list to store stemmed words
    y = []
     # Iterate through each word in the input text and append its stemmed form to the list
    for i in text.split():
        y.append(ps.stem(i))
        
    # Join the stemmed words into a single string and return
    return " ".join(y)
```

```python
# Apply the 'stem' function to the 'tags' column of the 'new_df' DataFrame
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Create a CountVectorizer instance with a maximum of 5000 features and English stop words removal
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
# Use the CountVectorizer to convert the 'tags' column into a matrix of token counts
vectors = cv.fit_transform(new_df['tags']).toarray()
```

# Now we are going to Calculate the cosine distance
 "Now, we are going to calculate the cosine distance between vectors. The choice of cosine distance over Euclidean distance is motivated by the fact that we are dealing with a high-dimensional space, specifically 5000 dimensions. In such high-dimensional spaces, cosine distance is often preferred due to its effectiveness in capturing similarity between vectors regardless of the dimensionality."

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Calculate the cosine similarity between vectors
similarity = cosine_similarity(vectors)
```

# Finding the distances and sorting

```python
# Sort and retrieve the indices of the top 5 most similar items to the first item in the 'similarity' matrix
# 'similarity[0]' represents the similarity scores for the first item
# The result is a list of tuples where each tuple contains the index and similarity score
# Sorting is done in descending order based on the similarity score
# The final list is sliced to exclude the first element (self-similarity) and keep the top 5 similar items

sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    
    # Find the index of the input movie in the DataFrame
    movie_index = new_df[new_df['title'] == movie].index[0]
    
    # Extract the cosine similarity distances for the input movie
    distances = similarity[movie_index]
    
    # Retrieve the indices of the top 5 similar movies based on cosine similarity
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    # Print the indices of the recommended movies
    for i in movies_list:
        print(i[0])
        
```
